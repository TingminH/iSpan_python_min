|string方法| 功能 |
|:---:|:---:|
|`str.replace('要被替換的字','要改成的字')`| 替換字串 |
|`str.stip()`| 去除兩側空格 |
|`str.upper()`| 字串小寫變大寫 |
|`str.lower()`| 字串大寫變小寫 |

```python
# 字串變數初始化
name = "Tingmin"
print(name)
print(name[0])

# string的排列[0,1,2,3]
```

```python
# 替換字串
# 方法 也是function的一種
# 定義在 class 底下的 function 可稱為 method
'''
str.replace()
將 string 中的 str1 替換成 str2
'''
name = 'Alex'
new_name = name.replace('ex','len') #遇到ex就替換成len
print(new_name)
```

```python
# 去除兩側空格
'''
str.strip()
去除字串 string 左、右兩邊的空格
'''

# 去除兩側空格
# str.strip()：去除字串 string 左、右兩邊的空格

data = "  |__ data __|  "
print(data)   # 原始字串

# 去除兩側空白
new_data = data.strip()
print(new_data)

# 去除所有空白
new_data = new_data.replace(' ', '')
print(new_data)

# 去除 |
new_data = new_data.replace('|', '')
print(new_data)

# 去除 _
new_data = new_data.replace('_', '')
print(new_data)
```

```python
# 字串小寫轉大寫 / 大寫轉小寫

# --- 小寫轉換 ---
# str.lower()：將字串 string 裡的字母全部改成小寫
data = "CAR"
new_data = data.lower()
print(new_data)   # 輸出 car

# --- 大寫轉換 ---
# str.upper()：將字串 string 裡的字母全部改成大寫
print("good".upper())   # 輸出 GOOD
```

```python
# 字串分割成 list
# str.split()

greeting = "hello world! hello John"
str_list = greeting.split() #預設是用空格去隔開!
print(str_list)
```
