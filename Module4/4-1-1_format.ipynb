|語法|說明|
|:---:|:---:|
|`%s` |插入 字串string  的值到字串中|
|`%f` |插入 浮點數float 的值到字串中|
|`%d` |插入 整數decimal integer的值到字串中|
|`%%` |輸出 百分比 (類似跳脫字元)|

```python
# 多組文字
# 老師的:
data1 = "Hello"
data2 = "World"
msg = '%s, %s! ' % (data1,data2)
print(msg)
```

```python
# 在新版 Python（3.6 之後），可以這樣寫，更簡單：
data1 = "Hello"
data2 = "World"
msg = f"{data1}, {data2}!"
print(msg)
```

```python
# 整數
age = 18
msg = 'I am %d years old.' %(age)
msf = '資料輸出的結果 Age = %d ' %(age)
print(msg)
print(age)
# 這樣可以直接知道哪一行出錯
```

```python
# 指定寬度 (維持 10 個字元長度，預設向右看齊)
msg = "[%10s]" % "apple"
print(msg)     # 結果: [     apple]
```

```python
# 靠左對齊 (維持 10 個字元長度，向左對齊)
msg = "[%-10s]" % "apple"
print(msg)     # 結果: [apple     ]
```

```python
# 靠右對齊（維持10個字元寬度，預設右對齊）
msg = "[%10s]" % "apple"
print(msg)      # [     apple]
```

```python
# 指定浮點數位數
# (保留小數點後 2 位)
msg = "[%.2f]" % 3.14159
print(msg)     # 結果: [3.14]
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
# 指定文字長度上限（只顯示前3個字母）
msg = "[%.3s]" % "pineapple"
print(msg)      # [pin]
```

```python
# 空白補 0
msg = "[%05d]" % 42
print(msg)      # [00042]
```
