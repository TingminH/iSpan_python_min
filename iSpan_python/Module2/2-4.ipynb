邏輯運算子 and / or / not

```python
name = 'Alex'
age = 18
if name == 'Alex' and age == 18:
    print('name: Alex; age: 18')
```

```python
name = 'Alex'
age = 20
if name == 'Alex' and age == 18:
    print('name: Alex; age: 18')
```

```python
name = 'Alex'
age = 20
if name == 'Alex' or age == 18:
    print('name: Alex; age: 18')
```

```python
name = 'Alex'
age = 20
if not(name == 'Alex' and age == 18): #if not可以用來防呆
    print('need to check')
```

成員運算子 in / not in

```python
num_list = [1, 2, 3, 4]

if 1 in num_list:
    print('1 in list')
```

```python
num_list = [1, 2, 3, 4]

if 5 not in num_list:
    print('5 not in list')
```

身分運算子 is / is not

```python
x = 10
y = 10
print(x)
print(y)
print(id(x)) #id:記憶體的位置
print(id(y)) #id:記憶體的位置
```

```python
if x is y:
    print('x和y參照到同一個記憶體位置')
```

```python
x = [1, 2, 3] 
y = [1, 2, 3]
if x == y:
    print('x 等於 y')

print(id(x)) #因為list可以改變 所以會用不同id
print(id(y))

if x is y:
    print('x和y參照到同一個記憶體位置')
```

位元運算子

```python
x = 1 # 0001 (位元運算值)
y = 3 # 0011

print(x&y) # 0001 # 1 (只顯示都有1的位元)
print(x|y) # 0011 # 1+2 (只要有1都顯示)
print(x^y) # 0010 # 2 (相同為0相異為1)
```
