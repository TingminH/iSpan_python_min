| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

##### for

```python

# Ctrl + D 選擇相同的文字做修改
# for <變數> in <可替代的物件>:
n = 1
# 實務上變數若為有意義的文字 通常會命名會直接用英文的單字
# for i in "contents": 這邊就不建議只是簡單用i來寫 以後回來看或跟同事討論會不知道是甚麼意思
for character in "contents":
    print(character, end='')
    print(n, end=',')
    n += 1
```

```python
# range(m) 走訪 0 到 m-1 的數字
for i in range(5):
    print(i)
```

```python
# range(n, m-1) 走訪 n 到 m-1 的數字
for i in range(1,11):
    print(i,end=',')

```

```python
# range(n, m-1, step) 以step為間距，走訪 n 到 m-1 的數字
for i in range(1,11,2):
    print(i,end=',')
```

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  
Output : 4 6 8 10 12

```python
def main():
    # 先設定a 跟 n
    a = int(input("'請輸入整數': "))
    n = int(input("'請輸入要產生幾個偶數': "))

    # 如果a是奇數 != 不等於
    if a % 2 != 0:
        a += 1
        for i in range(n):
            print(a + i * 2, end=" ")
    
    # 如果a是偶數
    else:
        for i in range(n):
            print(a + i * 2, end=" ")
        
main()
```

##### while

```python
count = 1

while count <= 5:
    print(count, end=',')
    # 要加上結束條件
    count += 1 
    

print ("Count is not < 5 now, loop ended.")
```
